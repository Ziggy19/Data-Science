# Matplotlibs

Pyplot
1. Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:



Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

# Markers
You can use the keyword argument marker to emphasize each point with a specified marker:

```python
y = np.array([3, 4, 7, 10])

plt.plot(y, marker = 'o')
plt.show()
```

```python
y = np.array([3, 4, 7, 10])

plt.plot(y, marker = '*')
plt.show()
```

# Format Strings fmt
You can use also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

marker|line|color

```python

ypoints = np.array([3, 2, 1, 5])

plt.plot(ypoints, 'o:r')
plt.show()
```

# Marker Size
You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:

```python
ypoints = np.array([3, 5, 4, 9])

plt.plot(ypoints, marker = 'o', ms = 10)
plt.show()
```

# Marker Color
You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:



```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dotted', marker = 'o', ms = 20, mec = 'r')
plt.show()
```
